```python
from pathlib import Path

import pandas as pd

ROOT = Path(__file__).resolve().parents[1]
ROBUSTNESS_DIR = ROOT / "results" / "week5" / "robustness"

kernel_path = ROBUSTNESS_DIR / "kernel_comparison.csv"

if not kernel_path.exists():
    raise FileNotFoundError(
        "Run python/scripts/run_week5_robustness.py to populate kernel_comparison.csv"
    )

metrics = pd.read_csv(kernel_path)
metrics.head()
```

```python
import matplotlib.pyplot as plt

try:
    import seaborn as sns
except ImportError as exc:
    raise ImportError("Install seaborn to render the heatmap") from exc

plt.rcParams.update({"figure.figsize": (8, 5)})

heatmap = metrics.loc[metrics["kernel"] == "exponential"]
pivot = heatmap.pivot_table(index="asset", columns="delta", values="rho")

sns.heatmap(pivot, annot=True, fmt=".3f", cmap="mako")
plt.title("Branching ratio ρ (exponential kernel)")
plt.xlabel("Δt (s)")
plt.ylabel("Asset")
plt.show()
```
